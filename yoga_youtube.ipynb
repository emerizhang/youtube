```python
#Wrap in functions
def get_videos(API_KEY, CHANNEL_ID):
    """Return df with video_id, video_title, upload_date, view_count, like_count, comment count

    Parameters:
    API_KEY: youtube API key
    CHANNEL_ID: channel id for channel of interest

    Makes Youtube API call to get videos for channel
    Calls get_video_details() to populate df with video details
    Returns: pandas df
     """
    #create an empty dataframe
    df = pd.DataFrame(columns=["video_id","video_title","upload_date","view_count","like_count","comment_count"]) 
    pageToken = ""
    while 1:
        url = "https://www.googleapis.com/youtube/v3/search?key="+API_KEY+"&channelId="+CHANNEL_ID+"&part=snippet,id&order=date&maxResults=10000&"+pageToken

        response = requests.get(url).json()
        time.sleep(1) #give it a second before starting the for loop to ensure the API call data is all grabbed
        for video in response['items']:
            if video['id']['kind'] == "youtube#video":
                video_id = video['id']['videoId']
                video_title = video['snippet']['title']
                video_title = str(video_title).replace("&amp;","")
                upload_date = video['snippet']['publishedAt']
                upload_date = str(upload_date).split("T")[0] #grab date: split on T, grab left hand
                view_count, like_count, comment_count = get_video_details(video_id)
                
                #append the info into dataframe
                df = df.append({'video_id':video_id,'video_title':video_title,
                                "upload_date":upload_date,"view_count":view_count,
                                "like_count":like_count,
                                "comment_count":comment_count},ignore_index=True)
        try:
            if response['nextPageToken'] != None: #if none, it means it reached the last page and break out of loop
                pageToken = "pageToken=" + response['nextPageToken']

        except:
            break

    return df
print(get_videos.__doc__)
```

```python
def get_video_details(video_id):
  """Makes Youtube API call to get stats for each video
  Parameters: 
  video_id: from previous search API call in get_videos()
  
  
  Returns: desired statistics
  Populates df with video details"""

  #collecting view, like, dislike, comment counts
  url_video_stats = "https://www.googleapis.com/youtube/v3/videos?id="+video_id+"&part=statistics&key="+API_KEY
  response_video_stats = requests.get(url_video_stats).json()

  view_count = response_video_stats['items'][0]['statistics']['viewCount']
  like_count = response_video_stats['items'][0]['statistics']['likeCount']
  comment_count = response_video_stats['items'][0]['statistics']['commentCount']

  return view_count, like_count, comment_count
print(get_video_details.__doc__)
```
